## Create random tensors

Rarely in practice you need to decide whether to use constant or variable, tf will automatically decide it for you. However, if in doubt,use constant and change it later if needed

Random tensor are filled with abitrary numbers

````python
random_t = tf.random.Generator.from_seed(42) # set seed for reprodutibility
r_1 = random_t.normal(shape=(3,3), dtype=np.float16)
r_1
r_2 = random_t.uniform(shape=(2,4), name="my random tensor")
r_2
````

````python
# lets shuffle a tensor

not_shuffled_t = tf.constant([
    [1,2,3],
    [4,3,2],
    [2,3,1]
])

not_shuffled_t
````

````python
# The tensor is shuffled only in his firt dimension, that means only on line if we have 2 dimensions

shuffled_t = tf.random.shuffle(value = not_shuffled_t)
shuffled_t
````

Tensor flow have few rules agains shuffle and radom seeds.
- check the [documentation](https://www.tensorflow.org/api_docs/python/tf/random/set_seed) about.

````python
# Reproducible shuffle
tf.random.set_seed(42) # global seed 42
tf.random.shuffle(value = not_shuffled_t, seed=42) # operation level seed equals to global
````

## Manipulating tensors

- Basic elementwise operation
- Tensor operations
    - add
    - sub
    - div
    - mult

````python
# You can add new values to a tensor using add operator
tensor = tf.zeros(shape = (3,3))

# basic four operations
tensor + 10 , tensor -10, (tensor+ 10) / 10, tensor + 2 * 10

# the original tensor is unchanged!
````

````python
# we can use builtin function to operat any tensor
# better to speedup on gpu,
# find more math operation on documentation

tf.add(tensor,10)
tf.multiply(tensor, 10)
tf.divide(tensor, 10)
tf.subtract(tensor, 10)
````

**Matrix operations**


````python
# Matrix multiplication
# the number of columns of the first op has to be equal
# of the number of lines of the second op

# mxn * mxn, m == n

vec_A = tf.constant(
    [1,2,3], shape=(3,1)
)
vec_A

mat_A = tf.constant([[1,2,3],
                    [5,6,6],
                    [5,6,7]
                     ])
mat_A
mat_B = tf.matmul(mat_A, vec_A)

mat_B
````

````python
# transpose a matrix
# reshape just shuffle the tensor around
# transpose shift axis

tf.transpose(mat_B) , tf.reshape(mat_B, shape=(3,1))
````

````python
transposed_A = tf.constant([[1,2,3],[4,5,6]])

transposed_A, tf.reshape(transposed_A , shape=(3,2)), tf.transpose(transposed_A)
````

````python
# Changing data type of a tensor

tensor = tf.constant([1,2,3,4])
tensor.dtype # default data type is int32

tf.cast(tensor, dtype=tf.float16)
````
